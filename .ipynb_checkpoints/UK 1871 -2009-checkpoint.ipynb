```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
import pylab
from pylab import *

from pandas import Series, DataFrame

%matplotlib inline

```

```python
#Load US and UK Inflation and Unemployment data
Maindata=pd.read_excel('UK US Inflation & Unemployment 1964 - 2009.xls')
Maindata=Maindata.rename(columns={'w':'Year','uki':'UK Inflation', 'usi':'US Inflation',
                         'usu':'US Unemployment','uku':'UK Unemployment'})
Maindata=Maindata.drop({'gi', 'gu'},1)
Maindata=Maindata.set_index('Year')
Maindata.head()

#Load Bank of England Data for GDP and Interest rates
BoE=pd.read_excel('BoE Data.xlsx')
BoE=BoE.set_index('Year')
BoE=BoE.rename(columns={'GDP':'UK GDP Change', 'Bank Rates':'UK Base Rate'})
BoE.head()

#Combine the data sets
Overall=pd.concat([BoE, Maindata], axis=1)

#Remove rows with any missing values
OverallUK=Overall.ix[:,['UK Base Rate', 'UK GDP Change',
                     'UK Inflation','UK Unemployment', 'Real GDP']]
OverallUK=OverallUK.dropna(axis=0, how='any')

#Overall=pd.concat([OverallUS, OverallUK], axis=1)
Overall=OverallUK.dropna(axis=0, how='any')

#Round each value to 1DP for uniformity and present table
Overall=Overall.round(1)
Overall.head()
```

```python
Overall.describe().round(1)
```

# Outliers

```python
sns.set_style("whitegrid")
box=sns.boxplot(Overall['UK Inflation'], orient='h')
box.set(xlabel='Inflation /%')
plt.savefig('Box Inflation w Outlier v2.jpg')
```

```python
Overall.ix[Overall['UK Inflation'] > 12, 'UK Inflation'] = Overall['UK Inflation'].mean()
Overall.ix[Overall['UK Inflation'] < -7, 'UK Inflation'] = Overall['UK Inflation'].mean()
```

```python
#Remove Outlier for UK
OverallUKI=OverallUK.ix[:,['UK Inflation']]

OverallInflation=OverallUKI.quantile([0.19, 0.89])
OverallUKI = OverallUKI.apply(lambda x: x[(x>OverallInflation.loc[0.19,x.name])
                                    & 
                                    (x < OverallInflation.loc[0.89,x.name])], 
                            axis=0)

OverallUK=OverallUK.drop({'UK Inflation'},1)
OverallUK=pd.concat([OverallUK, OverallUKI], axis=1)

OverallUK=OverallUK.dropna(axis=0, how='any')
```

```python
#Calculate R

Y=Overall['UK Inflation'].sum()
Y2= (Overall['UK Inflation'])**2
Y22=Y2.sum()
X=Overall['UK Base Rate'].sum()
X2=Overall['UK Base Rate']**2
X22=X2.sum()
n=Overall['UK Base Rate'].count()
XY=(Overall['UK Base Rate']*Overall['UK Inflation']).sum()

r=((n*XY)-(X*Y))/((((n*X22)-(X**2))*((n*Y22)-(Y**2)))**.5)
r
```

# MultiRegression

```python
X= Overall['UK Base Rate']
Y=Overall['UK Unemployment']
Z=Overall['UK Inflation']

data = pd.DataFrame({'x': X, 'y': Y, 'z': Z})

# Fit the model
model = ols("z ~ x + y", data).fit()

anova_results = anova_lm(model)
```

```python
Overall.loc[(Overall['UK Inflation']>=0) & (Overall['UK Inflation']<2.5), 'Group']='Steady'
Overall.loc[Overall['UK Inflation']>=2.5,'Group']='Weak'
Overall.loc[Overall['UK Inflation']<0,'Group']='Deflation'
Overall.loc[(Overall['UK Unemployment']>=8),'Group']='Disruptive'
```

```python
fig, ax = plt.subplots()
colors = {'Steady':'green', 'Weak':'blue', 'Disruptive': 'red', 'Deflation':'orange'}
grouped=Overall.groupby('Group')

for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='UK Unemployment', y='UK Inflation',
               label=key, color = colors[key], s=100)
    
plt.savefig('Clustering v3.jpg')
```
